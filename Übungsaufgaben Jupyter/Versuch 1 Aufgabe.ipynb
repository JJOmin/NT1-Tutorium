```python
# Import der benötigten Bibliotheken
import numpy as np
import matplotlib.pyplot as plt

# Aufgabe 1: Berechnung eines Sinussignals
def sin_signal(amplitude, frequency, time):
    """Erzeugt ein Sinussignal."""
    return amplitude * np.sin(2 * np.pi * frequency * time)

# Aufgabe 2: Berechnung der Pegelwerte
def pegel_value(amplitude):
    """Berechnet den Pegelwert in dBV."""
    return 20 * np.log10(amplitude / 1)

# Simulationsparameter
amplitude = 0.1  # Amplitude des Sinussignals in V
frequency = 1000  # Frequenz in Hz
time = np.linspace(0, 0.01, 1000)  # Zeitachse für die Simulation

# Berechnung des Sinussignals
signal = sin_signal(amplitude, frequency, time)

# Plot des Sinussignals
plt.figure(figsize=(10, 6))
plt.plot(time, signal)
plt.title("Sinussignal mit Amplitude = 0.1 V und Frequenz = 1 kHz")
plt.xlabel("Zeit (s)")
plt.ylabel("Amplitude (V)")
plt.grid(True)
plt.show()

# Aufgabe 3: Berechnung und Vergleich der Pegelwerte
theoretical_value = pegel_value(amplitude)
print(f"Theoretischer Pegelwert des Signals: {theoretical_value:.2f} dBV")

# Aufgabe 4: Rechtecksignal
def rect_signal(amplitude, frequency, duty_cycle, time):
    """Erzeugt ein Rechtecksignal mit einem gegebenen Tastverhältnis."""
    return amplitude * (signal > duty_cycle)

# Berechnung eines Rechtecksignals
duty_cycle = 0.5  # Tastverhältnis 50%
rect_signal = rect_signal(amplitude, frequency, duty_cycle, time)

# Plot des Rechtecksignals
plt.figure(figsize=(10, 6))
plt.plot(time, rect_signal)
plt.title("Rechtecksignal mit Amplitude = 0.1 V und Frequenz = 1 kHz")
plt.xlabel("Zeit (s)")
plt.ylabel("Amplitude (V)")
plt.grid(True)
plt.show()

# Aufgabe 5: Vergleich der Messergebnisse (Theorie vs. Messung)
measured_values = np.array([0.047, 0.033, 0.017, 0.00029, 0.0098, 0.011, 0.0066])
theoretical_values = np.array([0.05, 0.0335, 0.018, 0.0003, 0.0099, 0.0112, 0.0067])

# Plot zum Vergleich der Werte
plt.figure(figsize=(10, 6))
plt.plot(measured_values, label="Gemessen")
plt.plot(theoretical_values, label="Berechnet", linestyle='--')
plt.title("Vergleich der Messwerte: Rechtecksignal")
plt.xlabel("k-Werte")
plt.ylabel("Amplitude (V)")
plt.legend()
plt.grid(True)
plt.show()

# Aufgabe 6: Vergleich der Spektrumanalysatorwerte
spec_measured = np.array([0.042, 0.030, 0.014, 0.0003, 0.008, 0.01, 0.005])
spec_theoretical = np.array([0.043, 0.031, 0.015, 0.0003, 0.0085, 0.011, 0.0055])

# Plot zum Vergleich der Spektrumanalysatorwerte
plt.figure(figsize=(10, 6))
plt.plot(spec_measured, label="Spektrum Analysator (Gemessen)")
plt.plot(spec_theoretical, label="Spektrum Analysator (Berechnet)", linestyle='--')
plt.title("Vergleich der Spektrumanalysatorwerte")
plt.xlabel("k-Werte")
plt.ylabel("Amplitude (V)")
plt.legend()
plt.grid(True)
plt.show()
```
